```python
import numpy as np
```

```python
dict_data = {
    'edad': [10,9,13,14,12,11,12],
    'cm': [115,110,130,155,125,120,125],
    'pais':['co','mx','co','mx','mx','ch','ch'],
    'genero':['M','F','F','M','M','M','F'],
    'Q1':[5,10,8,np.nan,7,8,3],
    'Q2':[7,9,9,8,8,8,9]
}
dict_data
```

```python
#Convertir diccionario a dataframe, se agrega un index al dataframe
df= pd.DataFrame(dict_data, index = ['ana','benito','camilo','daniel','erika','fabian','gabriela'])
df
```

```python
df.values
```

```python
df['edad']
```

```python
df[['edad','cm','Q1']]
```

```python
#loc  = selecciona datos usando su index
df.loc['ana', ['edad','cm','Q1']]
```

```python
#iloc = selecciona datos usando su posicion en el dataframe
#df.iloc[2,1]
df.iloc[[2,4,5],[1,3]]
```

```python
# = indica que se requiere la info completa de toda la columna entera
df.iloc[:,[1,3]]
```

```python
#personas mayores a 12 aÃ±os
df[df['edad'] >= 12]
```

```python
df[(df['edad'] >= 12) & (df['pais'] == 'mx')]
```

```python
#estudiantes con calificacion mayor a la original
df[df['Q2'] >= df['Q2']]
```
