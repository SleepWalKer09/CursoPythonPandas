```python
import pandas as pd
import numpy as np
```

```python
pob = pd.read_csv('Poblacion.csv')
pob
```

```python
pob = pob.drop(columns=['Country Code','Indicator Name','Indicator Code','2019'])

pob = pd.melt(pob, id_vars=["Country Name"],
var_name="year", value_name="pob")

pob
```

```python
#formato legible para la poblacion
pd.options.display.float_format = '{:,.1f}'.format
pob
```

```python
#año converida a variable categorica para el manejo de multiples indices
pob['year'] = pd.Categorical(pob['year'].apply(str))
pob.dtypes
```

```python
#filtro para seleccion de paises
#isin == crea vector de variables booleanas
idxFiltro = pob['Country Name'].isin(['Aruba', 'Colombia'])
idxFiltro
```

```python
pobEjemplo = pob[idxFiltro]
pobEjemplo
```

```python
#nombre y año se convierten en indices
pobEjemplo = pobEjemplo.set_index(['Country Name', 'year']).sort_index()
pobEjemplo
```

```python
# seleccionar colombia
pobEjemplo.loc['Colombia', :]
```

```python
# seleccionar colombia y un año en particular
pobEjemplo.loc['Colombia', :].loc['2016',:]
```

```python
#anidar busqueda
pobEjemplo.xs(['Aruba'])
```

```python
#busqueda de multiples indices
pobEjemplo.xs(['Aruba', '2018'])
```

```python
#seleccion a bajo nivel
pobEjemplo.xs('2018', level='year')
```

```python
#ordenados A-Z por indices
paises = pob.set_index(['Country Name', 'year']).sort_index()
paises
```

```python
#ordenado Z-A por indices
paises = pob.set_index(['Country Name', 'year']).sort_index(ascending= [False, True])
paises
```

```python
#obtener multiples elementos del df
#elementos entre Aruba y Austria y que vayan del 2015 al 2017
ids = pd.IndexSlice
paises.loc[ids['Aruba': 'Austria', '2015':'2017'],:].sort_index()
```

```python
#primeros indices (nombre)
paises.index.get_level_values(0)
```

```python
#segundos indices (años)
paises.index.get_level_values(1)
```

```python
paises['pob']['Colombia']['2018']
```

```python
#suma de la poblacion mundial por año
paises.sum(level='year')
```

```python
#cambiar formato de indices
#convertir indices en columnas
pobEjemplo.unstack('Country Name')
```
