# Variable categorica: 
variables que definen una categoria, se reduce la memoria RAM que ocupan los archivos


```python
#En caso del dataset de meteoritos se tienen 2 variables categoricas, el nombre y la caida
meteorito.nunique()
```

```python
#conversion a variables categoricas
meteorito[['fall','nametype']] = meteorito[['fall','nametype']].astype('category')
```

```python
#verificacion de variables categoricas
meteorito.dtypes
```

```python
meteorito['fall'].unique()
```

```python
#encontrar cuantas veces se repite
meteorito['fall'].value_counts()
```

# variables dummies: Variables de gran importancia en el analisis de datos, la variable pasa de tener formato de texto a numerico


```python
pd.get_dummies(meteorito['fall'])
```

```python
meteorito['ones'] = 1
meteorito
```

```python
meteorito[['fell','found']] = pd.get_dummies(meteorito['fall'])
```

```python
meteorito['year'] #la fecha viene en tipo text, se necesita pasar a tipo fecha
```

# Borrar Filas, Columnas y copiar informacion


```python
meteorito.head(5)
```

```python
meteorito.drop(['ones'], axis= 1, inplace= True)
```

```python
#eliminar filas pares
meteorito.drop([0,2,4,6]).head(10)
```

```python
meteorito.drop(columns = ['id','recclass'], index=[0,2,4,6])
```

```python
#trabajar sobre una copia del dataframe, se debe especificar que la copia es "profunda", 
#la variable asignada (df) no copia directamente la info, si no copia la referencia a la memoria donde se guardan los datos
#si no se pone, es como seguir trabajando sobre el mismo dataframe
df = meteorito.copy(deep=True)
df.drop(['id'], axis=1, inplace=True)
```
