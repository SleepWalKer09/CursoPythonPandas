```python
import pandas as pd
```

```python
#definir 2 dataframe de listas
df_left = pd.DataFrame(
    {'X':['x0','x1','x2','x3'],
    'W':['w0','w1','w2','w3'],
    'Y':['y0','y1','y2','y3']},
    index = ['y2','y3','a2','a3'])

df_right = pd.DataFrame(
    {'Z':['z2','z3','z4','z5'],
     'A':['a2','a3','a4','a5'],
     'Y':['y2','y3','y4','y5']},
    index = [2,3,4,5])

```

```python
df_left
```

```python
df_right
```

```python
#merge: nuevo df con 
pd.merge(df_left,df_right)
```

```python
#especificar que tipo de union se esta haciendo
pd.merge(df_left,df_right, how = 'inner', on ='Y')
```

```python
#unir columna y con columna mix 
pd.merge(df_left,df_right, how='inner',left_on='Mix',right_on='Y')
```

```python
#mix se une a traves de Ã±a columna A
pd.merge(df_left,df_right, how='inner',left_on='Mix',right_on='A')
```

# Tipo de Merge

image.png

```python
# inner merge (vartiables en comun entre 2 dataframes)
pd.merge(df_left,df_right, how = 'inner', on ='Y')
```

```python
# left merge(df izquierdo se queda intacto y se le aÃ±aden los parametros del DF de la derecha)
pd.merge(df_left,df_right, how = 'left', on ='Y')
```

```python
# left merge (se respeta el tamaÃ±o y numeros del DF de la derecha y se le aÃ±aden los parametros de la izquierda)
pd.merge(df_left,df_right, how = 'right', on ='Y')
```

```python
#outer merge (unificar ambos dataframes mediante elementos en comun), imprime ambos dataframes
pd.merge(df_left,df_right, how = 'outer', on ='Y')

# pandas rellena con NaN cuando hay valores que no existan en el otro
```

# Merge para mas de un parametro

```python
#se crea una lista con los parametros para merge
#pandas por defecto lo realiza de esta manera, por eso es importante siempre decir que columnas estaran en el merge
pd.merge(df_left, df_right, how = 'outer')#, on =['Y','A'])
```

```python
#merge por sufijos
pd.merge(df_left, df_right, how = 'outer', on= 'Y', suffixes=['_left', '_right'])
```
