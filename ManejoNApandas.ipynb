```python
import pandas as pd
import numpy as np
```

```python
#Crear Dataframe
df = pd.DataFrame(np.arange(0, 15).reshape(5, 3), columns=['a', 'b', 'c'])
df
```

```python
# Añadir algunas variables no definidas:
df['d'] = np.nan
df['e'] = np.arange(15, 20)
df.loc[5,:] = pd.NA
df.loc[4,'a'] = pd.NA
df.loc[0,'d'] = 1
df.loc[5,'d'] = 10
df
```

```python
#Reconocer cuando un objeto es nulo simplmente usamos:
df.isnull()
```

# En dónde todas nuestras variables no definidas fueron marcadas con TRUE, df.isna() tambien cumple esta función.

```python
# Conocer el número de variables nulas por columna puede hacerse juntando el comando anterior con la funcion de suma:
df.isnull().sum()
```

```python
# Conocer el número de filas con elementos nulos:
df.notnull().sum(axis=1)
```

```python
# Todos los elementos nulos del dataframe
df.size-df.isnull().sum().sum()
```

### Reconocer estos elementos nos puede ayudar a filtrar en nuestro DataFrame

```python
# Filtrar por las variables no nulas de la columna 'a':
df[df['a'].notnull()]
```

```python
# Elimnar rapidamente las filas con registros faltantes:
df.dropna()
```

```python
df[['a']].dropna()
```

```python
#Usando la función fillna podremos reemplazarlas por el valor que querramos, en este caso 0.
df.fillna(0)
```

```python
#Remplazar con el valor siguiente usamos method="ffill":
df.fillna(method="ffill")
```

```python
#Remplazar con el valor previo usamos method="bfill":

df.fillna(method="bfill")
```

```python
#El mismo ejercicio anterior se puede aplicar con las filas usando axis=1:

df.fillna(method="bfill",axis=1)
```

```python
#Podemos usar también una serie para reemplazar los valores de una columna en especifico, es importante que haya emparejamiento entre los índices:

fill = pd.Series([100, 101, 102])
fill
```

```python
df['d'] = df['d'].fillna(fill)
df['d'] 
```

```python
df
```

```python
#Una de las formas más usadas para reemplazar datos es usar el promedio de las columnas, 
#esto se hace con la función mean. O si se quiere un mejor estimador, usamos median.

df.fillna(df.median())
```

```python
#Pandas también puede interporlar los valores faltanes calculando
#el valor que puede haber existido en el medio.

df_d = pd.concat([df[['d']], df[['d']].interpolate()],axis=1)
df_d.columns = ['d_antes','d_interpolado']
df_d
```
